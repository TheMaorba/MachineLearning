# Machine Learning


### ¿Que es el Machine learning?

El Machine Learning (aprendizaje automático) es una rama de la inteligencia artificial que permite a los sistemas aprender y mejorar automáticamente a partir de la experiencia, sin ser programados explícitamente para cada tarea. En lugar de seguir reglas fijas, los modelos identifican patrones en los datos y toman decisiones con mínima intervención humana.

### ¿Porque se usa Machine Learning?

Se usa porque permite resolver problemas donde las reglas tradicionales son demasiado complejas o imposibles de definir manualmente. Algunas razones principales son:

- Procesar grandes volúmenes de datos de forma eficiente.
- Detectar patrones ocultos que los humanos no podrían identificar fácilmente.
- Automatizar tareas repetitivas o de predicción.
- Mejorar con el tiempo a medida que recibe más datos.
- Aplicarse en áreas como medicina, finanzas, marketing, visión por computadora y procesamiento de lenguaje natural.

### ¿Que es una regresion lineal?

La regresión lineal es un modelo estadístico y de machine learning que busca describir la relación entre una variable dependiente (la que se quiere predecir) y una o más variables independientes mediante una ecuación lineal, ajustando una recta que minimiza el error entre los valores reales y los estimados, generalmente usando el método de mínimos cuadrados; en esencia, intenta encontrar la línea que mejor explica cómo cambian los datos.

### ¿Para que se usa una regresion lineal?

La regresión lineal se utiliza para predecir valores numéricos continuos y para analizar cuánto influye cada variable independiente en el resultado, siendo común en casos como estimar precios de viviendas, proyectar ventas, analizar salarios según experiencia o medir el impacto de la inversión en publicidad, ya que además de predecir, permite interpretar claramente la relación entre variables.

### ¿Que es una regresion logistica?

La regresión logística es un modelo de machine learning supervisado que se utiliza para problemas de clasificación y que, aunque su nombre diga “regresión”, no predice valores continuos sino probabilidades; funciona combinando las variables independientes en una expresión lineal y luego aplicando una función sigmoide para transformar ese resultado en un valor entre 0 y 1, el cual se interpreta como la probabilidad de que una observación pertenezca a una clase determinada.

### ¿Para que se utiliza una regresion logistica?

La regresión logística se utiliza para predecir categorías, especialmente en problemas de clasificación binaria como detectar fraude, aprobar o rechazar un crédito, diagnosticar una enfermedad o predecir si un cliente abandonará un servicio, ya que permite estimar probabilidades, establecer umbrales de decisión y además interpretar el impacto de cada variable en términos de aumento o disminución del riesgo.

# Cross- Validation

### ¿Que es el Cross-Validation?

El Cross-Validation (validación cruzada) es una técnica de evaluación de modelos de machine learning que consiste en dividir el conjunto de datos en varias partes (o "folds") para entrenar y evaluar el modelo de manera iterativa, permitiendo obtener una estimación más robusta del rendimiento del modelo al reducir la variabilidad causada por la división aleatoria de los datos.




# Hyperparameter Tuning

### ¿Que es el Hyperparameter Tuning?

El Hyperparameter Tuning (ajuste de hiperparámetros) es el proceso de optimizar los hiperparámetros de un modelo de machine learning, que son los parámetros que se establecen antes del entrenamiento y que no se aprenden a partir de los datos, con el objetivo de mejorar el rendimiento del modelo; esto se puede hacer mediante técnicas como la búsqueda en cuadrícula (Grid Search), la búsqueda aleatoria (Random Search) o métodos más avanzados como la optimización bayesiana.

# Métricas de Evaluación de Regresión

### ¿Cuales son las métricas de evaluación de regresión?
- Error Cuadrático Medio (MSE) - Mean Squared Error
- Error Absoluto Medio (MAE) - Mean Absolute Error
- R² (Coeficiente de Determinación) - R-squared

### ¿Que es el Error Cuadrático Medio (MSE) - Mean Squared Error?

Es una métrica que mide la media de los cuadrados de los errores, es decir, la diferencia entre los valores reales y los valores predichos al cuadrado. Es sensible a los outliers porque eleva al cuadrado los errores.

### ¿Para que se usa el Error Cuadrático Medio (MSE) - Mean Squared Error?
Se usa para evaluar el rendimiento de modelos de regresión, especialmente cuando se quiere penalizar más los errores grandes, ya que al elevar al cuadrado, los errores más grandes tienen un impacto mayor en la métrica, lo que puede ser útil en situaciones donde es importante evitar predicciones muy inexactas.
Ejemplo: En la predicción de precios de viviendas, un error grande podría significar una diferencia significativa en el costo, por lo que el MSE ayudaría a identificar y penalizar esas predicciones inexactas.

### ¿Que es elError Absoluto Medio (MAE) - Mean Absolute Error?

Es una métrica que mide la media de los valores absolutos de los errores, es decir, la diferencia entre los valores reales y los valores predichos sin elevar al cuadrado. Es menos sensible a los outliers que el MSE.

### ¿Para que se usa el Error Absoluto Medio (MAE) - Mean Absolute Error?
Se usa para evaluar el rendimiento de modelos de regresión cuando se quiere tener una medida más interpretable de los errores, ya que el MAE representa la media de las diferencias absolutas entre las predicciones y los valores reales, lo que facilita entender cuánto se está equivocando el modelo en promedio, sin penalizar excesivamente los errores grandes como lo hace el MSE.
Ejemplo: En la predicción de la demanda de productos, un error de 10 unidades podría ser aceptable, y el MAE ayudaría a entender el error promedio sin exagerar el impacto de errores más grandes.

### ¿Que es el R² (Coeficiente de Determinación) - R-squared?

Es una métrica que indica qué proporción de la variabilidad total en la variable dependiente puede explicarse por las variables independientes del modelo. Su valor está entre 0 y 1, donde 1 indica que el modelo explica toda la variabilidad.

### ¿Para que se usa el R² (Coeficiente de Determinación) - R-squared?
Se usa para evaluar el rendimiento de modelos de regresión, ya que proporciona una medida de qué tan bien el modelo se ajusta a los datos, indicando la proporción de la variabilidad en la variable dependiente que es explicada por el modelo, lo que ayuda a entender la capacidad predictiva del modelo.
Ejemplo: En la predicción de precios de automóviles, un R² de 0.8 indicaría que el modelo explica el 80% de la variabilidad en los precios, lo que sugiere un buen ajuste a los datos.
