# Machine Learning


### ¿Que es el Machine learning?

El Machine Learning (aprendizaje automático) es una rama de la inteligencia artificial que permite a los sistemas aprender y mejorar automáticamente a partir de la experiencia, sin ser programados explícitamente para cada tarea. En lugar de seguir reglas fijas, los modelos identifican patrones en los datos y toman decisiones con mínima intervención humana.

### ¿Porque se usa Machine Learning?

Se usa porque permite resolver problemas donde las reglas tradicionales son demasiado complejas o imposibles de definir manualmente. Algunas razones principales son:

- Procesar grandes volúmenes de datos de forma eficiente.
- Detectar patrones ocultos que los humanos no podrían identificar fácilmente.
- Automatizar tareas repetitivas o de predicción.
- Mejorar con el tiempo a medida que recibe más datos.
- Aplicarse en áreas como medicina, finanzas, marketing, visión por computadora y procesamiento de lenguaje natural.

### ¿Que tecnicas de Machine Learning existen?

Aprendizaje Supervisado

El modelo aprende de datos etiquetados.

- Regresión Lineal y Logística
- Árboles de Decisión y Random Forest
- Support Vector Machines (SVM)
- Redes Neuronales
- Gradient Boosting (XGBoost, LightGBM)

Aprendizaje No Supervisado

El modelo encuentra patrones sin etiquetas.


- Clustering (K-Means, DBSCAN, Jerárquico)
- Reducción de dimensionalidad (PCA, t-SNE)
- Detección de anomalías
- Reglas de asociación (Apriori)


Aprendizaje por Refuerzo

El modelo aprende mediante recompensas y penalizaciones.

- Q-Learning
- Deep Q-Network (DQN)
- Policy Gradient

### ¿Que es una regresion lineal?

La regresión lineal es un modelo estadístico y de machine learning que busca describir la relación entre una variable dependiente (la que se quiere predecir) y una o más variables independientes mediante una ecuación lineal, ajustando una recta que minimiza el error entre los valores reales y los estimados, generalmente usando el método de mínimos cuadrados; en esencia, intenta encontrar la línea que mejor explica cómo cambian los datos.

### ¿Para que se usa una regresion lineal?

La regresión lineal se utiliza para predecir valores numéricos continuos y para analizar cuánto influye cada variable independiente en el resultado, siendo común en casos como estimar precios de viviendas, proyectar ventas, analizar salarios según experiencia o medir el impacto de la inversión en publicidad, ya que además de predecir, permite interpretar claramente la relación entre variables.

### ¿Que es una regresion logistica?

La regresión logística es un modelo de machine learning supervisado que se utiliza para problemas de clasificación y que, aunque su nombre diga “regresión”, no predice valores continuos sino probabilidades; funciona combinando las variables independientes en una expresión lineal y luego aplicando una función sigmoide para transformar ese resultado en un valor entre 0 y 1, el cual se interpreta como la probabilidad de que una observación pertenezca a una clase determinada.

### ¿Para que se utiliza una regresion logistica?

La regresión logística se utiliza para predecir categorías, especialmente en problemas de clasificación binaria como detectar fraude, aprobar o rechazar un crédito, diagnosticar una enfermedad o predecir si un cliente abandonará un servicio, ya que permite estimar probabilidades, establecer umbrales de decisión y además interpretar el impacto de cada variable en términos de aumento o disminución del riesgo.
